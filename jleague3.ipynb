```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as ply
%matplotlib inline
data_t_raw=pd.read_excel('trainのコピー.xls')
data_val=pd.read_excel('testのコピー.xls')
data_c_raw=pd.read_excel('conditionのコピー.xls')
data_t2=data_t_raw.copy(deep=True)
data_c2=data_c_raw.copy(deep=True)
data_t2.sort_values(by='id',inplace=True)
data_c2.sort_values(by='id',inplace=True)
data_2=pd.merge(data_t2,data_c2,how='inner')
```

```python
data_2_J1=data_2[data_2.stage=='Ｊ１']
data_2_J2=data_2[data_2.stage=='Ｊ２']
data_2_J1.reset_index(drop=True,inplace=True)
data_2_J2.reset_index(drop=True,inplace=True)
```

```python
import datetime
time1=datetime.time(9,0,0)
time2=datetime.time(15,0,0)
time3=datetime.time(18,0,0)
time4=datetime.time(23,59,59)
```

```python
for i in range(len(data_2_J1['time'])):
    if time1<= data_2_J1['time'][i]<time2:
         data_2_J1['time'][i]=0
    elif time2<= data_2_J1['time'][i]<time3:
         data_2_J1['time'][i]=1
    elif time3<=  data_2_J1['time'][i]<time4:
         data_2_J1['time'][i]=2
    else:
         data_2_J1['time'][i]=3
            
tbin=pd.get_dummies(data_2_J1['time'],columns=['tbin1','tbin2','tbin3'])
tbin.columns=['tbinJ1_1','tbinJ2_2','tbinJ3_3']
tbin
```

```python
data_2_J1_month=data_2_J1['gameday'].copy(deep=True)
```

```python
for i in range(len(data_2_J1)):
    if '土' in data_2_J1['gameday'][i]:
         data_2_J1['gameday'][i]='土'
    elif '日' in data_2_J1['gameday'][i]:
         data_2_J1['gameday'][i]='日'
    elif '祝' in data_2_J1['gameday'][i]:
         data_2_J1['gameday'][i]='祝'
    else:
         data_2_J1['gameday'][i]='平日'
```

```python
data_2_J1_holiday=pd.get_dummies(data_2_J1['gameday'])
```

```python
for j in range(1,13):
    for i in range(len(data_2_J1)):
        if type(data_2_J1_month[i])==int:
            data_2_J1_month[i]+=0
        elif int(data_2_J1_month.str.split('/')[i][0])==j:
            data_2_J1_month[i]=j
```

```python
data_2_J1_m=pd.get_dummies(data_2_J1_month)
```

```python
data_2_J1_2=data_2[data_2.stage=='Ｊ１'].copy(deep=True)
data_2_J1_2.reset_index(drop=True,inplace=True)
```

```python
for i in range(len(data_2_J1_2)):
    if '雨' in data_2_J1_2['weather'][i]:
        data_2_J1_2['weather'][i]='雨'
    else:
        data_2_J1_2['weather'][i]='雨でない'
        
```

```python
data_2_weather_dum=pd.get_dummies(data_2_J1_2['weather'])
```

```python
data_2_J1_sta_dum=pd.get_dummies(data_2_J1['stadium'])
```

```python
data_2_J1_all=pd.concat([data_2_J1['id'],data_2_weather_dum,data_2_J1_sta_dum,tbin,data_2_J1_holiday,data_2_J1_m],axis=1)
```

```python
from sklearn import svm, tree, linear_model, neighbors, naive_bayes, ensemble, discriminant_analysis, gaussian_process
from sklearn import model_selection
MLA = [
    #Ensemble Methods
    #ensemble.GradientBoostingClassifier(),
    ensemble.RandomForestClassifier(),
    
    #GLM
    #linear_model.LogisticRegressionCV(),
    #linear_model.RidgeClassifierCV(),
    
    #SVM
    #svm.SVC(probability=True),
    ]



#split dataset in cross-validation with this splitter class: http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.ShuffleSplit.html#sklearn.model_selection.ShuffleSplit
#note: this is an alternative to train_test_split
cv_split = model_selection.ShuffleSplit(n_splits = 10, test_size = .3, train_size = .6, random_state = 0 ) # run model 10x with 60/30 split intentionally leaving out 10%

#create table to compare MLA metrics
MLA_columns = ['MLA Name','MLA Train Accuracy Mean', 'MLA Test Accuracy Mean']
MLA_compare = pd.DataFrame(columns = MLA_columns)

#create table to compare MLA predictions
MLA_predict = data_2_J1_all

#index through MLA and save performance to table
row_index = 0
for alg in MLA:

    #set name and parameters
    MLA_name = alg.__class__.__name__
    MLA_compare.loc[row_index, 'MLA Name'] = MLA_name
    
    #score model with cross validation: http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_validate.html#sklearn.model_selection.cross_validate
    cv_results = model_selection.cross_validate(alg, data_2_J1_all, data_2_J1['y'], cv  = cv_split)

    MLA_compare.loc[row_index, 'MLA Time'] = cv_results['fit_time'].mean()
    MLA_compare.loc[row_index, 'MLA Train Accuracy Mean'] = cv_results['train_score'].mean()
    MLA_compare.loc[row_index, 'MLA Test Accuracy Mean'] = cv_results['test_score'].mean()   
    #if this is a non-bias random sample, then +/-3 standard deviations (std) from the mean, should statistically capture 99.7% of the subsets


    #save MLA predictions - see section 6 for usage
    alg.fit(data_2_J1_all, data_2_J1['y'])
    MLA_predict[MLA_name] = alg.predict(data_2_J1_all)
    
    row_index+=1

    
#print and sort table: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sort_values.html
MLA_compare.sort_values(by = ['MLA Test Accuracy Mean'], ascending = False, inplace = True)
MLA_compare
#MLA_predict
```
